```python
link_dicts = []
for idx, row in sampled_cid_data.iterrows():
    link_dict = {
        "selstate": row["STATE_FIPS"],
        "selcounty": row["CNTY_FIPS"],
        "selcommunity": row["CID"],
        "searchedCid": row["CID"]
    }
    link_dicts.append(link_dict)
```

```python
base_dict = {
        "utf8": "âœ“",
        "affiliate": "fema",
        "query": "",
        "selstate": "",
        "selcounty": "",
        "selcommunity": "",
        "jurisdictionkey": "",
        "jurisdictionvalue": "",
        "searchedCid": "",  # Same as selcommunity
        "searchedDateStart": "",
        "searchedDateEnd": "",
        "txtstartdate": "",
        "txtenddate": "",
        "method": "search"
    }
```

```python
link_comms_df = pd.DataFrame(link_dicts)
```

```python
params_dicts = []
for idx, row in link_comms_df.iterrows():
    params = base_dict.copy()
    params['selstate'] = row['selstate']
    params['selcounty'] = row['selcounty']
    params['selcommunity'] = row['selcommunity']
    params['searchedCid'] = row['searchedCid']
    params_dicts.append(params)
```
